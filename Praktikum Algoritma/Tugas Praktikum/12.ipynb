```python
class AlatTulis:
    def __init__(self, nama, stok, harga_satuan):
        self.nama = nama
        self.stok = stok
        self.harga_satuan = harga_satuan
        self.harga_total = stok * harga_satuan

    def set_nama(self, nama):
        self.nama = nama

    def get_nama(self):
        return self.nama

    def get_stok(self):
        return self.stok

    def get_harga_satuan(self):
        return self.harga_satuan

    def get_total_harga(self):
        return self.harga_total

class Transaction:
    def __init__(self):
        self.total_harga = 0

    def set_total_harga(self, alat_tulis_list):
        for alat_tulis in alat_tulis_list:
            self.total_harga += alat_tulis.get_total_harga()

    def get_total_harga(self):
        return self.total_harga

# Membuat objek AlatTulis
bolpoint = AlatTulis("Bolpoint", 10, 2000)
pensil = AlatTulis("Pensil", 10, 1000)
penghapus = AlatTulis("Penghapus", 10, 500)

# Mengatur nama AlatTulis
bolpoint.set_nama("Bolpoint Baru")

# Menampilkan informasi setiap AlatTulis
for alat_tulis in [bolpoint, pensil, penghapus]:
    print("\nNama\t\t:", alat_tulis.get_nama())
    print("Stok\t\t:", alat_tulis.get_stok())
    print("Harga Satuan\t:", alat_tulis.get_harga_satuan())
    print("Total Harga\t:", alat_tulis.get_total_harga())

# Membuat objek Transaction
transaction = Transaction()

# Melakukan transaksi
transaction.set_total_harga([bolpoint, pensil, penghapus])

# Menampilkan total harga setelah transaksi
print("\nTotal Harga setelah transaksi adalah:", transaction.get_total_harga())
```
